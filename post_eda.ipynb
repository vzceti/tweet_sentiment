```python
# Run the imports/functions file
%run ./imports_functions.ipynb

# Read in data and create tweets list
df = pd.read_csv("predicting_data.csv")
tweets = splitTweets(df.tweet_text.tolist())

# Create dictionaries for word cloud (key is the word, value is the frequency)
tweet_dict = {}
for i in range(len(tweets.unique_values)):
  if tweets.iloc[i].unique_values[0] == 'x' or tweets.iloc[i].unique_values in list(string.ascii_letters):
    continue
  tweet_dict[tweets.iloc[i].unique_values] = tweets.iloc[i].counts
  
# Read in the mask shape and set stopwords
twitter = np.array(Image.open("twitter.png"))
stopwords = set(STOPWORDS)

# Set wordcloud parameters and save to file
# Display disabled for presentation
wc = WordCloud(stopwords=stopwords, font_path = "VeraMono.ttf", background_color="black", mask = twitter, max_words = 300, colormap = "Blues").fit_words(tweet_dict)
wc.to_file('tweet_cloud.png')
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.show()
```
